```python
def GaussianKernel(size, sigma):
    arr = np.arange(-size // 2 + 1, size // 2 + 1)
    x, y = np.meshgrid(arr, arr)
    kernel = np.exp(-(x**2 + y**2) / (2 * sigma ** 2)) 
    kernel /= kernel.sum()
    return np.array(kernel)
```
